```python
import matplotlib.pyplot as plt
import numpy as np
import glob
import os
```

```python
import matplotlib.cm as mplcm
import matplotlib.colors as colors
NUM_COLORS = 12

cm = plt.get_cmap('gist_rainbow')
cNorm  = colors.Normalize(vmin=0, vmax=NUM_COLORS-1)
scalarMap = mplcm.ScalarMappable(norm=cNorm, cmap=cm)
color_bar = [scalarMap.to_rgba(i) for i in range(NUM_COLORS)]
```

```python
path_chart = glob.glob('model/FS_017/*.npy')
path_chart
```

```python
#i = 0
plt.figure(figsize=(14, 14))
plt.title('Comparison of 12 set')
plt.xlim((51, 100))
plt.ylim((0.5, 1))
plt.axhline(y=0.7, color='black', linestyle='-')
for i in range(len(path_chart)):
    train_chart, valid_chart = np.load(path_chart[i])
    serial = os.path.basename(path_chart[i]).split('_')[1].split('.')[0]
    plt.plot(train_chart, label = f'{serial} - t', color = color_bar[i])
    plt.plot(valid_chart, label = f'{serial} - v', color = color_bar[i] , linestyle=':')
plt.legend(loc=2)
plt.show()
```
